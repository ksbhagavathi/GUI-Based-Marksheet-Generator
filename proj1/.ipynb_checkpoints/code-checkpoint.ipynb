```python
class Student:
    
    def __init__(self,roll,name):
        self.roll=roll
        self.name=name
        self.details=None
        self.right=0
        self.wrong=0
        self.notans=0
        self.marks=0
```

```python
import pandas as pd
import numpy as np
import os
import csv 
import xlsxwriter
from openpyxl.reader.excel import load_workbook
from openpyxl.styles import*
import PySimpleGUI as sg
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders
```

```python
def set_color(ws,rows,cols,rs,cs,s):
    
    
    rfont=Font(name="Century", size=12,color='FF0000')
    bfont=Font(name="Century", size=12,color='0000FF')
    gfont=Font(name="Century", size=12,color='008000')
    
    
    for row in range(1,rows):
        cell=ws[row+rs][cs]
        ccell=ws[row+rs][cs+1]
        if cell.value==ccell.value:
            cell.font=gfont   
        else:
            cell.font=rfont
        ccell.font=bfont
        
    
        
    align1= Alignment(horizontal="right",vertical="center")       
    ws['A7'].alignment=align1
    ws['A8'].alignment=align1
    ws['D7'].alignment=align1
```

```python
def makeworkbook(r,students):
     
    s=students[r]   
    output_file=r+'.xlsx'
    
    if os.path.exists(output_file):
        os.remove(output_file)
        
    wb=xlsxwriter.Workbook(output_file)
    ws=wb.add_worksheet()

   
    ws.set_column(0,10,18)
    
    ws.insert_image('A1','logo.jpeg')
    
    headf=wb.add_format({'bold':True,'font_size':20, 'font_name':'Century','underline':True})
    
    
    ws.write('C6','Mark Sheet',headf)
    
    mainf=wb.add_format({'bold':True,'font_size':12, 'font_name':'Century'})
    cf=wb.add_format({'font_size':12, 'font_name':'Century'})
    ws.write('A7','Name :',cf)
    ws.write('B7',s.name,mainf)
    ws.write('A8','Roll Number :',cf)
    ws.write('B8',r,mainf)
    ws.write('D7','Exam :',cf)
    ws.write('E7','quiz',mainf)
    
    
    
    wb.close()
    
```

```python

def set_border(ws, rows,cols,rs,cs):
    
    border = Border(left=Side(border_style='medium', color='000000'),
                right=Side(border_style='medium', color='000000'),
                top=Side(border_style='medium', color='000000'),
                bottom=Side(border_style='medium', color='000000'))

    for row in range(rows):
        for col in range(cols):
            cell=ws[row+rs][col+cs]
            cell.border = border
            
```

```python
def set_prop(ws,arr,rs):
    
    align= Alignment(horizontal="center",vertical="center")
    font=Font(name="Century", size=12)
    
    for i in range(len(arr)):
        for j in range(len(arr[i])):
            cell=ws[i+rs][j]
            cell.alignment=align
            cell.font=font
    
    for i in range(len(arr)):
        cell=ws[i+rs][0]
        cell.font=Font(name="Century", size=12,bold=True)
        
    for j in range(len(arr[0])):
        cell=ws[rs][j]
        cell.font=Font(name="Century", size=12,bold=True)
            
            
```

```python
def addtofile(ws,r,arr,k):

    for i in range(len(arr)):
        for j in range(len(arr[i])):
            ws.cell(row=i+k,column=j+1,value=arr[i][j])
            
    set_border(ws,len(arr),len(arr[0]),k,0)
    set_prop(ws,arr,k)       
      
```

```python
def answersdf(students,r,c):
    
    wb=load_workbook(r+'.xlsx')
    ws=wb.active
    
    ansbox=[]
    s=students[r].details
    
    ansbox.append(['Student Ans', 'Correct Ans'])
    
    for i in range(len(c)):
        ansbox.append([s[i],c[i]])
    
    addtofile(ws,r,ansbox,16)
    
    set_color(ws,len(c)+1,2,16,0,students[r])
    
    wb.save(r+'.xlsx')
    wb.close()
    
   
```

```python
def makeresultdf(s,c,mr,mw,mn):
    
    print(s.right,s.marks)
    v=[]

    v.append(['','Right','Wrong','Not Attempt','Max'])
    v.append(['No',s.right,s.wrong,s.notans,len(c)])
    v.append(['Marking',mr,mw,mn,''])
    s.marks=s.right*mr+s.wrong*mw+s.notans*mn
    v.append(['Total',s.right*mr,s.wrong*mw,'',str(s.marks)+'/'+str(len(c)*mr)])
    
    return v
```

```python
def resultdf(s,r,c,mr,mw,mn):
    
    wb=load_workbook(r+'.xlsx')
    ws=wb.active
    
    data=makeresultdf(s,c,mr,mw,mn)
    addtofile(ws,r,data,10)
    
    for i in range(1,len(data)):
        cell1=ws[10+i][1]
        cell2=ws[10+i][2]
        
        cell1.font=Font(name="Century", size=12,color='008000')
        cell2.font=Font(name="Century", size=12,color='FF0000')
    
    ws['E13'].font=Font(name="Century", size=12,color='0000FF')
        
        
        
    wb.save(r+'.xlsx')
    wb.close()
```

```python
def find_marks(s,c):
    k=s.details
    s.right=0
    s.notans=0
    s.wrong=0
    for i in range(len(c)):
        if k[i]==c[i]:
            s.right=s.right+1
        elif type(k[i])==float:
            s.notans=s.notans+1
        else:
            s.wrong=s.wrong+1
    
```

```python
def Generate_marksheet(students,mr,mw,c):
    
    for r in students:
        if r!='Answer':
            s=students[r]
            find_marks(s,c)
            makeworkbook(r,students)
            answersdf(students,r,c)
            resultdf(s,r,c,mr,mw,0)
            
                 
    return 1
```

```python
def GenerateConciseMarksheet(students,c,mfile,mr,mn,mw):
    
    tmp4=[]
    tmpl=[]
    
    mf=pd.read_csv(mfile)
    for index,x in mf.iterrows():

        r=x['Roll Number']
        s=students[r]
        find_marks(s,c)
        students[r].marks=s.right*mr+s.wrong*mw+s.notans*mn
        tmp4.append(str(students[r].marks)+'/'+str(students['ANSWER'].marks))
        dd=[students[r].right,students[r].wrong,students[r].notans]
        tmpl.append(dd)


    mf.insert(4, 'Score_After_Negative', tmp4)
    mf = mf.assign(statusAns=tmpl)
    
    ofile='concise_marksheet.csv'
    if os.path.exists(ofile):
        os.remove(ofile)
    mf.to_csv(ofile)
```

```python
def read_inputs(values,flag1,flag2,flag3):
    
    students=dict()
    
    mr=float(values['mr'])
    mw=float(values['mw'])
    
    file=values['master_roll_file']  
    mfile=values['responses_file']

    df=pd.read_csv(file)
    for index,x in df.iterrows():
        if x['roll'] not in students:
            students[x['roll']]=Student(x['roll'],x['name']) 

    mf=pd.read_csv(mfile)

    for index,x in mf.iterrows():   
        students[x['Roll Number']].details=x[7:]
    
    if 'ANSWER' not in students:
        print('error')
        return -1

    c=students['ANSWER'].details  
    
    if flag1==1:
        GenerateConciseMarksheet(students,c,mfile,mr,0,mw)
    if flag2==1:
        Generate_marksheet(students,mr,mw,c)
    if flag3==1:
        #send_emails(students,c)
        print(1)
    
    return 0
        
        
    
```

```python
sg.theme('Dark2')

layout = [[sg.Text("master_roll file in .csv format: "), 
           sg.Input(key='master_roll_file') ,sg.FileBrowse()],
          [sg.Text("responses file in .csv format: "), 
           sg.Input(key='responses_file'),sg.FileBrowse()],
          [sg.Text('marks for correct answer:'),sg.Input(key='mr')],
          [sg.Text('marks for wrong answer:'),sg.Input(key='mw')],
          [sg.Button('Generate concise marksheet')],
          [sg.Button('Generate Roll number wise marksheet')],
          [sg.Text(size=(25,1), k='-OUTPUT-')],
          [sg.Button('Send emails')],
          [sg.Button('Exit')]
         ]

window = sg.Window('MarkSheet Generator', layout)


while True:
    
    event, values = window.read()

    #print(event, values)

    if event in (None, 'Exit'):
        window.close()
        break
    elif event=='Generate concise marksheet':
        try :
            checker=read_inputs(values,1,0,0)
            if checker==-1:
                sg.popup('responses file did not have ANSWER ')
                break 
        except:
            sg.popup('error in code')
            break    
        
    elif event=='Generate Roll number wise marksheet':
        
        try :
            checker=read_inputs(values,0,1,0)
            if checker==-1:
                sg.popup('responses file did not have ANSWER ')
                break 
        except:
            sg.popup('error in code')
            break
        
    elif event=='Send emails':
        
        try :
            checker=read_inputs(values,0,1,1)
            if checker==-1:
                sg.popup('responses file did not have ANSWER ')
                break 
        except:
            sg.popup('error in code')
            break  
        
    
window.close()

```
